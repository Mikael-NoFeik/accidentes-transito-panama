## Tipos de Variables

```python
tipos_variables = {
    "Mes": "Cualitativa ordinal",
    "Año": "Cuantitativa discreta",
    "Accidentes": "Cuantitativa discreta",
    "Temperatura Prom. del Mes (°C)": "Cuantitativa continua",
    "Acum. Lluvia Diaria": "Cuantitativa continua",
    "Días Hábiles del mes": "Cuantitativa discreta",
    "Inicio de Clases": "Dicotómica",
    "Precio Gasolina (95 Octanos)": "Cuantitativa continua",
    "Accidentes del mes anterior": "Cuantitativa discreta",
    "IMAE": "Cuantitativa continua"
}

for variable, tipo in tipos_variables.items():
    print(f"{variable}: {tipo}")
```

## Medidas de Dispersión
Las medidas de dispersión permiten evaluar qué tan variables son los datos con respecto a una medida de tendencia central. En este apartado se calcularán el rango, la varianza y la desviación estándar de las principales variables numéricas del conjunto de datos.

Las variables analizadas son:

- Accidentes
- Temperatura promedio del mes
- Acumulación de lluvia diaria
- Días hábiles del mes
- Precio de la gasolina (95 octanos)
- IMAE

Posteriormente, se realizará un análisis más detallado de la variable **Accidentes**, ya que constituye la variable de estudio principal para el desarrollo del modelo de regresión lineal múltiple.

```python
variables = [
    "Accidentes",
    "Temperatura Prom. del Mes (°C)",
    "Acum. Lluvia Diaria",
    "Días Hábiles del mes",
    "Precio Gasolina (95 Octanos)",
    "IMAE"
]

for variable in variables:
    print("="*60)
    print(variable)
    print("="*60)

    print(f"Rango: {datos[variable].max()-datos[variable].min():.2f}")
    print(f"Varianza: {datos[variable].var():.2f}")
    print(f"Desviación estándar: {datos[variable].std():.2f}")

    print()
```

### Análisis de la variable Accidentes
Debido a que la variable **Accidentes** es la variable dependiente del proyecto y será utilizada posteriormente en el modelo de regresión lineal múltiple, se realiza un análisis descriptivo más detallado.

```python
print("Variable: Accidentes")
print("="*60)

print(f"Media: {datos['Accidentes'].mean():.2f}")
print(f"Mediana: {datos['Accidentes'].median():.2f}")

print("\nModa:")
print(datos["Accidentes"].mode())

print(f"\nMínimo: {datos['Accidentes'].min()}")
print(f"Máximo: {datos['Accidentes'].max()}")

print(f"Rango: {datos['Accidentes'].max()-datos['Accidentes'].min():.2f}")
print(f"Varianza: {datos['Accidentes'].var():.2f}")
print(f"Desviación estándar: {datos['Accidentes'].std():.2f}")
```

## Cuartiles

Los cuartiles dividen los datos en cuatro pates iguales. se presentan el primer cuartil (Q1), la mediana (Q2) y el tercer cuartil (Q3) para las variables seleccionadas.

```python
variables = [
    "Accidentes",
    "Temperatura Prom. del Mes (°C)",
    "Acum. Lluvia Diaria",
    "Días Hábiles del mes",
    "Precio Gasolina (95 Octanos)",
    "IMAE"
]

for variable in variables:
    print("="*60)
    print(variable)
    print("="*60)
    print(datos[variable].quantile([0.25, 0.50, 0.75]))
    print()
```
